## Stock Sentiment Analysis using News Headlines

```python
import pandas as pd
df=pd.read_csv('D:\Supratim2\Code\Python\Predict Stock Price Movement Based On News Headline using NLP\Data.csv', encoding = "ISO-8859-1")
df.head(1)
```

```python
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']
```

```python
X_train=train.iloc[:,2:27]
X_test=test.iloc[:,2:27]
y_train=train.iloc[:,1]
y_test=test.iloc[:,1]
```

```python
# Remove HTML tags
import pandas as pd
import re
def remove_html_tags(text):
    if isinstance(text, str):
        return re.sub(r'<.*?>', '', text)
    else:
        return text
X_train = X_train.applymap(remove_html_tags)
```

```python
# Removing punctuations
import pandas as pd
import string
def remove_punctuation(text):
    if isinstance(text, str):
        return text.translate(str.maketrans('', '', string.punctuation))
    else:
        return str(text)  # Convert non-string to string or handle differently
# Apply the function to each element in the DataFrame, handling non-string values
X_train = X_train.applymap(remove_punctuation)
```

```python
# Lowercasing
X_train = X_train.apply(lambda col: col.str.lower() if col.dtype == 'object' else col)
```

```python
# Function to remove URLs
import re
def remove_urls(text):
    return re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
X_train = X_train.apply(lambda col: col.apply(remove_urls) if col.dtype == 'object' else col)
```

```python
# Remove stopwords using NLTK
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))  # Set of English stopwords
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)
X_train = X_train.apply(lambda col: col.apply(remove_stopwords) if col.dtype == 'object' else col)
```

```python
# Remove emojis using regex
import re
def remove_emojis(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U0001F700-\U0001F77F"  # alchemical symbols
                               u"\U00002600-\U000027BF"  # miscellaneous symbols
                               u"\U0001F780-\U0001F7FF"  # black/white chess symbols
                               u"\U000024C2-\U0001F251"  # enclosed characters
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
X_train = X_train.apply(lambda col: col.apply(remove_emojis) if col.dtype == 'object' else col)
```

```python
# Stemming on text using NLTK
import nltk
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
nltk.download('punkt') 
porter = PorterStemmer()
def stem_text(text):
    tokens = word_tokenize(text)
    stemmed_tokens = [porter.stem(token) for token in tokens]
    return ' '.join(stemmed_tokens)
X_train = X_train.apply(lambda col: col.apply(stem_text) if col.dtype == 'object' else col)
```

```python
# Lemmatization on text using NLTK
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
nltk.download('punkt')  # Download the Punkt tokenizer models if not already downloaded
nltk.download('wordnet')  # Download WordNet if not already downloaded
lemmatizer = WordNetLemmatizer()
def lemmatize_text(text):
    tokens = word_tokenize(text)
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return ' '.join(lemmatized_tokens)
X_train = X_train.apply(lambda col: col.apply(lemmatize_text) if col.dtype == 'object' else col)
```

```python
X_train_headline = []
for row in range(0,len(X_train.index)):
    X_train_headline.append(' '.join(str(x) for x in X_train.iloc[row,:]))
```

```python
X_test = X_test.applymap(remove_html_tags)
X_test = X_test.applymap(remove_punctuation)
X_test = X_test.apply(lambda col: col.str.lower() if col.dtype == 'object' else col)
X_test = X_test.apply(lambda col: col.apply(remove_urls) if col.dtype == 'object' else col)
X_test = X_test.apply(lambda col: col.apply(chatword_treatment) if col.dtype == 'object' else col)
# X_test = X_test.apply(lambda col: col.apply(correct_spelling) if col.dtype == 'object' else col)
X_test = X_test.apply(lambda col: col.apply(remove_stopwords) if col.dtype == 'object' else col)
X_test = X_test.apply(lambda col: col.apply(remove_emojis) if col.dtype == 'object' else col)
X_test = X_test.apply(lambda col: col.apply(stem_text) if col.dtype == 'object' else col)
X_test = X_test.apply(lambda col: col.apply(lemmatize_text) if col.dtype == 'object' else col)
```
